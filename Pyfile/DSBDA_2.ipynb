```python
sorted_rscore = sorted(dt['Reading_Score'])
sorted_rscore
q1 = np.percentile(sorted_rscore,25)
q3 = np.percentile(sorted_rscore,75)
IQR = q3-q1
```

```python
lwrBound = q1 - (0.5*IQR)
upperBound = q3 + (0.5*IQR)
```

```python
outliers = []
for i in dt['Reading_Score']:
    if i<lwrBound or i>upperBound:
        outliers.append(i)

outl = np.where(dt['Reading_Score']<lwrBound )
print(outliers)
dt.iloc[outl] 
# outl
```
